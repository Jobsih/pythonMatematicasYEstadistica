```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,3)
fig,ax=plt.subplots()
y=x**2+3
l1=ax.plot(x,y, '-', color= 'blue')
ax.scatter(0.5,0, color='red')
ax.grid(True)
indice=0
for i in x[10:40]:
    m=2*i
    ax.plot(x,m*x-(m*x[indice])+y[indice],'-', color="orange")
    indice+=1
```

## Clase 4

```python
rnd= np.random.default_rng(4567)
datos= np.arange(15)
print(datos)
probabilidades = np.array(
    [0.3, 0.2, 0.1, 0.05, .05, 0.05, 0.05, 0.025,
    0.025, 0.025, 0.025, 0.025, 0.025, 0.025, 0.025]
)

assert round(sum(probabilidades),10) ==1.0, "La probabilidad debe sumar 1"
seleccion= rnd.choice(datos,p= probabilidades, replace=True)
print(seleccion)
seleccion_datos= rnd.choice(datos, p=probabilidades, replace=True, size=(4,3))
print(seleccion_datos)
```

```python
rndDos= np.random.default_rng(2456)
rndFlotantes= rndDos.random(size=(3,3))
print(rndFlotantes)

rndF= np.random.uniform(1.2,1.60,size=10)
print(rndF)

rndenteros=rndDos.integers(5,15,size=8)
print(rndenteros)

distribucion = rndDos.integers(1,10,size=20)
fig,ax=plt.subplots()
ax.hist(distribucion,linewidth=0.5, edgecolor="white")
ax.set_title("Histograma de aleatorios")
ax.set_xlabel("Valor")
ax.set_ylabel("Frecuencia")
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
plt.show()
```

```python
rnd=np.random.default_rng(45)
mu=5.0
sigma=3
rnds=rnd.normal(loc=mu,scale=sigma, size=1000)

fig,ax=plt.subplots()
ax.hist(rnds,bins=20,linewidth=0.5, edgecolor="white")
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
plt.show()
```
