### Ejercicio 1

```python
def f1(x):
    return (3*x-6)/2
def f2(x):
    return (6*x+9)/4
x=np.linspace(-.5,3)
y1=f1(x)
y2=f2(x)
lines= plt.plot(x,y1 , x,y2)
```

```python
def f1(x):
    return x-3
def f2(x):
    return (x+1)/-4
x=np.linspace(-.5,5)
y1=f1(x)
y2=f2(x)
lines= plt.plot(x,y1 , x,y2)
```

```python
def f1(x,y):
    return (3+(-2*x)+y)/2
def f2(x,y):
    return ((-2*x)+(-3*y))/-1
def f3(x,y):
    return (-12-x+(-2*y))/2
#inicio,fin
x=np.linspace(5,-5)
y=np.linspace(5,-5)
X,Y=np.meshgrid(x,y)
Z1=f1(X,Y)
Z2=f2(X,Y)
Z3=f3(X,Y)

ax = plt.figure().add_subplot(111, projection="3d") 

ax.plot_surface(X, Y, Z1, alpha = 0.5, rstride = 100, cstride = 100)
ax.plot_surface(X, Y, Z2, alpha = 0.5, rstride = 100, cstride = 100)
ax.plot_surface(X, Y, Z3, alpha = 0.5, rstride = 100, cstride = 100)
```
