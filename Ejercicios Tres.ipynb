### Ejercicio 1

```python
def f1(x):
    return (3*x-6)/2
def f2(x):
    return (6*x+9)/4
x=np.linspace(-.5,3)
y1=f1(x)
y2=f2(x)
lines= plt.plot(x,y1 , x,y2)
```

```python
def f1(x):
    return x-3
def f2(x):
    return (x+1)/-4
x=np.linspace(-.5,5)
y1=f1(x)
y2=f2(x)
lines= plt.plot(x,y1 , x,y2)
```

```python
def f1(x,z):
    return 2*x+2*z-3
def f2(x,z):
    return (2*x-z)/3
def f3(x,z):
    return (x+2*z+12)/2
inicio=-10
fin= 10
#inicio,fin
x=np.linspace(inicio,fin)
z=np.linspace(inicio,fin)
y1=f1(x,z)
y2=f2(x,z)
y3=f3(x,z)

ax = plt.figure().add_subplot(projection='3d')
ax.plot(x,y1,z, label="f1", color='red')
ax.plot(x,y2,z, label="f2", color='blue')
ax.plot(x,y3,z, label="f3", color='green')
```
